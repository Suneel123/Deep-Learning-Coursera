# Planar Data Classification with One Hidden Layer
+ Implement a  2-class classification neural network with a single hidden layer
+ Use hidden units with a non-linear activation such as Tanh
+ Compute cross-entropy loss
+ Implement forward and backward propagation

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
import sklearn.linear_model

%matplotlib inline
np.random.seed(1)
```

```python
def load_planar_dataset():
    np.random.seed(1)
    m = 400 # number of random examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m, D)) # data matrix where each row is a single example
    Y = np.zeros((m, 1), dtype='uint8') # labels vector 0 for red 1 for blue
    a = 4 # maximum ray of flower
    
    for j in range(2):
        ix = range(N*j, N*(j+1))
        t = np.linspace(j*3.12, (j+1)*3.12, N) + np.random.randn(N)*0.2  # theta
        r = a*np.cos(4*t) + np.random.randn(N)*0.2 # radius
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        print(X[ix].shape)
        Y[ix] = j
        
    X = X.T
    Y = Y.T
    return X, Y
```

```python
X, Y = load_planar_dataset()
```

```python
def load_panar_dataset1(n=8, d=4):
    a = 4  # scale
    n = float(n); d = float(d)
    k = n / d  # number of petals (k) if k is odd, double number of petals (2k) if k is even 
    X = np.zeros((4000, 2))
    ix = range(0, 4000)
    t = np.linspace(0, 80, 4000)
    r = a*np.cos(k*t)
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    return X.T
```

## Simple Logistic Regression

```python
clf = sklearn.linear_model.LogisticRegressionCV(cv=5)
clf.fit(X.T, Y.T.ravel());
```

```python
def plot_decision_boundary(model, x, y):
    # Set minimum and maximum values and give it some padding
    x_min, x_max = x[0, :].min() - 1, x[0, :].max() + 1
    y_min, y_max = x[1, :].min() - 1, x[1, :].max() + 1
    h = 0.01
    # Generate grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # plot the contour of training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(x[0, :], x[1, :], c=y.ravel(), cmap=plt.cm.Spectral)
```

```python
# plot decision boundary for logistic regression
plot_decision_boundary(lambda x: clf.predict(x), X, Y)
plt.title("Logistic Regression")

# Print accuracy
LR_predictions = clf.predict(X.T) 
print("Accuracy of the logistic regression: {} (percentage of correctly labelled datapoints)".format(float(np.dot(Y, LR_predictions) + 
                                                                                                           np.dot(1-Y, 1-LR_predictions))/
                                                                                                    float(Y.size)*100))
```

## Neural Network Model
+ Logistic regression did not work well, we will use Neural Network to train the model with a single hidden layer.

Mathematically, 

For one example $x^{(i)}$
$$Z^{[1](i)} = W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$<br/> 
$$a^{[1](i)} = \tanh(Z^{[1](i)})\tag{2}$$<br/> 
$$Z^{[2](i)} = W^{[2]} a^{[i](i)} + b^{[2]}\tag{3}$$<br/> 
$$\hat{y}^{(i)} = a^{[2](i)} = \sigma(Z^{[2](i)})\tag{4}$$<br/>
$$\hat{y}^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$<br/>

Given the predictions on all the examples, you can compute the cost as:
$$J = -\frac{1}{m} \sum_{i=1}^{m} \large\left(\small y^{(i)}\log\left(a^{[2](i)}\right) + (1-y^{(i)})\log\left(1-a^{[2](i)}\right) \large \right) \small \tag{6}$$

**Reminder**: The general methodology to build a Neural Network is to:
1. Define the neural network structure ( # of input units,  # of hidden units, etc). 
2. Initialize the model's parameters
3. Loop:
    1. Implement forward propagation
    2. Compute loss
    3. Implement backward propagation to get the gradients
    4. Update parameters (gradient descent)

You often build helper functions to compute steps 1-3 and then merge them into one function we call `nn_model()`. Once you've built `nn_model()` and learnt the right parameters, you can make predictions on new data.

### Defining the Neural Network Structure
Define three variables:
1. n_x: the size of the input layer
2. n_h: the size of the hidden layer (set this to 4) 
3. n_y: the size of the output layer

Hint: Use shapes of X and Y to find n_x and n_y. Also, hard code the hidden layer size to be 4.

```python
def layer_sizes(X, Y):
    n_x = X.shape[0]
    n_h = 4
    n_y = Y.shape[0]
    return (n_x, n_h, n_y)
```

### Initialize the models parameters

```python
def initialize_parameters(n_x, n_h, n_y):
    """
    Initializes random weights and biases for the network
    
    Parameters:
    -----------
    n_x: size of input layer
    
    n_h: size of hidden layer
    
    n_y: size of output layer
    
    Return:
    -------
    params: dictionary containing
        W1 - weight matrix of shape (n_h, n_x)
        b1 - bias vector of shape (n_h, 1)
        W2 - weight matrix of shape (n_y, n_h)
        b2 - bias vector of shape (n_y, 1) 
    """
    np.random.seed(2)
    
    W1 = np.random.randn(n_h, n_x)*0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y, n_h)*0.01
    b2 = np.zeros((n_y, 1))
    
    assert (W1.shape == (n_h, n_x))
    assert (b1.shape == (n_h, 1))
    assert (W2.shape == (n_y, n_h))
    assert (b2.shape == (n_y, 1))
    
    parameters = {'W1': W1,
                  'b1': b1,
                  'W2': W2,
                  'b2': b2} 
    
    return parameters
```

We have defined the neural network, and initialized weights and biases, now finally we need to loop through computing forward propagation, cost function, backward propagation, updating weights and biases  

Forward Propagation:

+ Retrieve each parameter from the dictionary, implement foraward propagation. Compute $Z^{[1]}, A^{[1]}, Z^{[2]}, A^{[2]}$ (the vector of all your predictions on all the examples in the training set).
+ Values needed for back propagation are stored in 'cache'. The 'cache' will be given as input to backpropagation function. 

```python
def sigmoid(Z):
    """
    Computes sigmoid function for the input Z
    
    Arguments:
    ----------
    Z: Input to sigmoid function
    
    Returns:
    --------
    A: sigmmoid(Z)
    """
    A = 1. / (1 + np.exp(-Z))
    return A
```

```python
def forward_propagation(X, parameters):
    """
    Compute steps in forward propagation
    
    Arguments:
    -----------
    X: input to neural network of shape (n_x, m)
    parameters: dict, contains initialization parameters
    
    Returns:
    --------
    A2: the sigmoid output of second activation function
    cache: dict, containing Z1, A1, Z2, A2 
    """
    
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    # forward propagation
    Z1 = np.dot(W1, X) + b1      # W1:(n_h, n_x) * X:(n_x, m) -> (n_h, m) + b1:(n_h, 1) -> Z1:(n_h, m)
    A1 = np.tanh(Z1)             # A1:(n_h, m)
    Z2 = np.dot(W2, A1) + b2     # W2:(n_y, n_h) * A1:(n_h, m) -> (n_y, m) + b2:(n_y, 1) -> Z1:(n_y, m)
    A2 = sigmoid(Z2)             # A2:(n_y, m)
    
    cache = {'Z1': Z1,
             'A1': A1, 
             'Z2': Z2, 
             'A2': A2}
    
    return A2, cache
```

```python
def compute_cost(A2, Y, parameters):
    """
    Compute cross-entropy cost given in eq 6.
    
    Arguments:
    ----------
    A2: Activation of the output layer (n_y, m)
    Y: Actual outputs for the input training examples (1, m)
    parameters: dict, containing weights and biases
    
    Returns:
    --------
    cost: cross-entropy cost given in equation 6.
    """
    m = Y.shape[1]
    cost = (-1. / m) * np.sum(Y*np.log(A2) + (1-Y)*np.log(1-A2))
    cost = np.squeeze(cost)
    return cost
```

### Gradient Descent

$$dZ^{[2]} = A^{[2]} - Y\tag{7}$$

$$dW^{[2]} = \frac{1}{m}dZ^{[2]} A^{[1]^{T}}\tag{8}$$

$$db^{[2]} = \frac{1}{m}\text{ }np.sum(dZ^{[2]},\text{ }axis=1,\text{ }keepdims=True)\tag{9}$$

$$dZ^{[1]} = W^{[2]^{T}} dZ^{[2]} * g^{[1]^{'}}(Z^{[1]})\tag{10}$$

$$dW^{[1]} = \frac{1}{m}dZ^{[1]} X^{T}\tag{11}$$

$$db^{[1]} = \frac{1}{m}\text{ }np.sum(dZ^{[1]},\text{ }axis=1,\text{ }keepdims=True)\tag{12}$$

Note:
+ To compute dZ1 you'll need to compute $g^{[1]'}(Z^{[1]})$. Since $g^{[1]}(.)$ is the tanh activation function = $\dfrac{e^{z} - e^{-z}}{e^{z} + e^{-z}}$, if $a = g^{[1]}(z)$ then $g^{[1]'}(z) = 1-a^2$. So you can compute 
    $g^{[1]'}(Z^{[1]})$ using `(1 - np.power(A1, 2))`.

```python
# back-propagation
def backward_propagation(parameters, cache, X, Y):
    """
    Compute backward propagation
    
    Arguments:
    ----------
    parameters: dict, containing our parameters
    cacahe: dict, containing 'Z1', 'A1', 'Z2', and 'A2'
    X: (n_x, m), input data of shape, (2, number of examples) 
    Y: (n_y, m), 'true' labels vector of shape (1, number of examples)
    
    Returns:
    --------
    grads: dict, containing gradients with respect to different parameters
    """
    m = X.shape[1]
    
    W1 = parameters['W1']   # W1: (n_h, n_x) -> (no. of hidden layers, no. of input layers)
    W2 = parameters['W2']   # W2: (n_y, n_h) -> (no. of output layers, no. of hidden layers)
    
    A1 = cache['A1']   # A1: (n_h, m) -> (no. of hidden layers, no. of training examples)
    A2 = cache['A2']   # A2: (n_y, m) -> (no. of output layers, no. of training examples)
    
    dZ2 = A2 - Y    # dZ2: (n_y, m)
    dW2 = (1./m) * np.dot(dZ2, A1.T)  # dW2: (n_y, n_h)
    db2 = (1./m) * np.sum(dZ2, axis=1, keepdims=True)  # db2: (n_y, 1) 
    dZ1 = np.dot(W2.T, dZ2) * (1 - np.power(A1, 2))  # dZ1: (n_h, m)
    dW1 = (1./m) * np.dot(dZ1, X.T)  # dW1: (n_h, n_x)
    db1 = (1./m) * np.sum(dZ1, axis=1, keepdims=True)
    
    grads = {'dW2': dW2,
             'db2': db2,
             'dW1': dW1,
             'db1': db1}
    
    return grads
```

### Update parameters
Use gradient descent to update the parameters. Gradient descent rule, 

$\theta' = \theta - \alpha \dfrac{\partial J}{\partial \theta}$<br>
where,<br> 
&nbsp;&nbsp;&nbsp;&nbsp;$\alpha$ = learning rate <br>
&nbsp;&nbsp;&nbsp;&nbsp;$\theta$ = parameter

```python
# update paramters
def update_paramters(parameters, grads, learning_rate=1.2):
    """
    Update paramters using gradient descent rule
    
    Arguments:
    ----------
    paramters: dict, containing parameters
    grads: dict, containing gradients
    
    Returns:
    --------
    parameters: dict, containing updated paramters
    """
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]
    
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

## Neural Network model

```python
# Neural network model
def nn_model(X, Y, n_h, num_iterations=10000, print_cost=False):
    """
    Arguments:
    ----------
    X: (2, m) input 
    Y: (1, m) output
    n_h: no. of hidden layers
    num_iterations: no. of iterations in gradient descent loop
    print_cost: if True, print the cost of every 1000 iterations
    
    Returns:
    --------
    parameters: prameters learnt by model, that can be used to predict.
    """
    
    np.random.seed(3)
    n_x, _, n_y = layer_sizes(X, Y)
    
    # initialize parameters
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    # iterating through neural network
    for i in range(num_iterations):
        # forward propagation
        A2, cache = forward_propagation(X, parameters)
        
        # compute cost
        cost = compute_cost(A2, Y, parameters)
        
        # backpropagation
        grads = backward_propagation(parameters, cache, X, Y)
        
        # update parameters
        parameters = update_paramters(parameters, grads)
            
        if print_cost and i%1000 == 0:    
            print("Cost after iteration {}: {}".format(i, cost))
    return parameters
```

## Predictions
$y_{prediction} = \begin{cases}
    1 & \text{if}\ activation > 0.5 \\
    0 & \text{otherwise}
    \end{cases}$

```python
def predict(paramters, X):
    """
    Using learned parameters, predicts a class for each example in X
    
    Arguments:
    ----------
    parameters: dict, containing learned parameters
    X: (n_x, m), input data
    
    Returns:
    --------
    predictions: a vector of predictions, (red: 0 / blue: 1)
    """
    A2, _ = forward_propagation(X, paramters)
    predictions = A2 > 0.5
    return predictions
```

Run the model and see how it performs on a planar dataset. Test the model, with single hidden layer of n_h units

```python
%%time
# Build NN model
parameters = nn_model(X, Y, n_h=4, print_cost=True)

# plot decision boundary
plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
plt.title('Decision boundary for hidden layer of size: 4')
```

```python
# Print accuracy
predictions = predict(parameters, X)
print ('Accuracy: %d' % float((np.dot(Y,predictions.T) + np.dot(1-Y,1-predictions.T))/float(Y.size)*100) + '%')
```

## Tuning hidden layer size


```python
%%time
plt.figure(figsize=(16, 32))
hidden_layer_sizes = [1, 2, 3, 4, 5, 20, 50]
for i, n_h in enumerate(hidden_layer_sizes):
    plt.subplot(5, 2, i+1)
    plt.title(f'Hidden layer of size: {n_h}')
    parameters = nn_model(X, Y, n_h, num_iterations=5000)
    plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
    predictions = predict(parameters, X)
    accuracy = float((np.dot(Y,predictions.T) + np.dot(1-Y,1-predictions.T))/float(Y.size)*100)
    print ("Accuracy for {} hidden units: {} %".format(n_h, accuracy))
```

### Interpretation
+ The larger models (with more hidden units) are able to fit the training set better, until eventually the largest models overfit the data.
+ The best hidden layer size seems to be around n_h = 20. Indeed, a value around here seems to fits the data well without also incurring noticable overfitting.
+ We will take a look at regularization, which lets you use very large models (such as n_h = 50) without much overfitting.
